```python
import pandas as pd
from datetime import datetime as dt, timedelta
import matplotlib.pyplot as plt
import json

groups = [{"group_id":-1001877079626, "group_name":"DeepOcean FX (Member)"}, {"group_id":-1001781939396, "group_name":"DeepOcean TH Stock"}
          , {"group_id":-1001967606094, "group_name":"DeepOcean TH Stock (Premium)"}]

for group in groups:
    df = pd.read_csv(f"csv/inout_member_hist{group['group_id']}.csv")

    if len(df.index) > 0:
        fig, ax = plt.subplots()

        date = [ str(x) for x in df["date"]]
        new_mem_df = [ json.loads(x) for x in df["in"]]
        counts = [ len(json.loads(x)) for x in df["in"]]

        left_mem_df = [ json.loads(x) for x in df["out"]]
        
        ax.bar(date, counts)
        print(f"New/Left member(s) of group:{group['group_name']}")
        ax.set_ylabel(f'In/Out member {group["group_name"]}')
        ax.set_title(f'In/Out member {group["group_name"]}')
        # ax.legend(title=f'In/Out member {group["group_name"]}')

        print("="*20)
        print(f"New members==>")
        no = 1
        for member in new_mem_df:
            if member['user_id']:
                key = list(member['user_id'].keys())
                print(f"\t{no}. Member Id:{member['user_id'][key[0]]}, Member Username:{member['user_name'][key[0]]}\n")
            no += 1
        
        no = 1
        print(f"Left the group members==>")
        for member in left_mem_df:
            if member['user_id']:
                key = list(member['user_id'].keys())
                print(f"\t{no}.Member Id:{member['user_id'][key[0]]}, Member Username:{member['user_name'][key[0]]}\n")
            no += 1
        plt.show()
        
        
```
